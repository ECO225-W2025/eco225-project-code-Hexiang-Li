```python
def clean_listings(df):
    # Drop duplicates
    df = df.drop_duplicates()

    # Drop irrelevant or redundant columns (example: URLs, host names, etc.)
    columns_to_drop = [
        'listing_url', 'scrape_id', 'last_scraped', 'picture_url',
        'host_url', 'host_name', 'host_picture_url'
    ]
    df = df.drop(columns=[col for col in columns_to_drop if col in df.columns], errors='ignore')

    # Handle missing values (example: fill numeric NaNs with median, categorical with mode)
    for col in df.columns:
        if df[col].dtype in ['int64', 'float64']:
            df[col] = df[col].fillna(df[col].median())
        elif df[col].dtype == 'object':
            df[col] = df[col].fillna(df[col].mode()[0])

    return df
```

```python
listings_cleaned = clean_listings(listings)
```

```python
# Cleaning the Calendar Dataset
def clean_calendar(df):
    # Drop duplicates
    df = df.drop_duplicates()

    # Convert date column to datetime
    if 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'], errors='coerce')

    # Convert price column to numeric (removing symbols like $)
    if 'price' in df.columns:
        df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)

    # Handle missing values
    df = df.fillna(method='ffill').fillna(method='bfill')

    return df
```
